```
import pandas as pd
```

```
# 데이터 파일을 업로드한 경로
path = '/content/drive/My Drive/Colab Notebooks/202007/gapminder.tsv'
# 파일을 읽어서 DataFrame 생성
gapminder = pd.read_csv(path, sep='\t')
gapminder
```

```
# csv(comma-separated values) 파일
# 데이터들이 구분자(separator)로 구별되어 있는 텍스트 파일
# 구분자: 쉼표(,), 탭('\t'), 콜론(':', '::'), ...
```

```
# DF에서 컬럼 선택: dataframe[컬럼], dataframe[[컬럼1, 컬럼2, ...]]
# gapminder DF에서 국가이름, 1인당GDP 컬럼 선택
gapminder[['country', 'gdpPercap']]
```

```
# DF에서 행(row)를 선택: dataframe.loc[], dataframe.iloc[]
gapminder.loc[0:4]  # gapminder.head()
```

```
gapminder.iloc[0:5]
```

```
# gapminder DF에서 마지막 행 5개를 출력: gapminder.tail()
gapminder.iloc[-5:]  # 음수 인덱스는 iloc만 사용 가능.
```

```
# boolean indexing: 
# dataframe[조건식]: DF에서 조건식을 참으로 만드는 row들을 검색
# 2007년 데이터 검색
gapminder[gapminder['year'] == 2007]
```

```
# boolean indexing과 컬럼 선택 조합
# 2007년의 국가이름, 인구, 1인당GDP 검색
gapminder[gapminder['year'] == 2007][['country', 'pop', 'gdpPercap']]
```

```
gapminder_2007 = gapminder[gapminder['year'] == 2007]
cols = ['country', 'pop', 'gdpPercap']
gapminder_2007[cols]
```

```
# Afghanistan 데이터만 검색
gapminder[gapminder['country'] == 'Afghanistan']
```

```
gapminder[gapminder['country'] == 'Korea, Rep.']
```

```
gapminder[gapminder['country'].str.contains('Korea')]
```

```
# dataframe[(조건식1) & (조건식2)]
# &(and), |(or)
# Asia, 2007년 자료 검색
gapminder[(gapminder['continent'] == 'Asia') & 
          (gapminder['year'] == 2007)]
```

```
# 평균(mean), 최댓값(max), 최솟값(min), 표준편차(std), 중앙값(median), ...
gapminder['gdpPercap'].median()
```

```
# 1인당GDP가 최댓값인 레코드(row)를 검색
gdp_max = gapminder['gdpPercap'].max()
gapminder[gapminder['gdpPercap'] == gdp_max]
```

```
# 1인당GDP의 평균보다 큰 gdp를 갖는 레코드들을 검색
gpd_mean = gapminder['gdpPercap'].mean()
gapminder[gapminder['gdpPercap'] > gpd_mean]
```

```
# 대륙별 1인당GDP 평균
gapminder.groupby('continent')['gdpPercap'].mean()
```

```
# 연도별 1인당GDP의 평균
gapminder.groupby('year')['gdpPercap'].mean()
```

```
# 연도별, 대륙별 1인당GDP 평균
df_year_continent = \
    gapminder.groupby(['year', 'continent'])['gdpPercap'].mean()
df_year_continent
```

```
df_unstacked = df_year_continent.unstack()
df_unstacked
```

```
import matplotlib.pyplot as plt

plt.plot(df_unstacked['Africa'], label='Africa')
plt.plot(df_unstacked['Americas'], label='America')
plt.plot(df_unstacked['Asia'], label='Asia')

plt.legend()  # 범례(label) 보여주기
plt.xlabel('Year')
plt.ylabel('GDP per capita')
plt.title('GDP per capita vs Year')
plt.show()
```

```
# 연도별, 대륙별 인구 합(sum)
df_year_continent2 = \
    gapminder.groupby(['year', 'continent'])['pop'].sum()
df_unstack2 = df_year_continent2.unstack()
df_unstack2
```

```
# plt.plot(df_unstack2['Africa'], label='Africa')
# plt.plot(df_unstack2['Americas'], label='Americas')
# plt.plot(df_unstack2['Asia'], label='Asia')

for continent in df_unstack2.columns:
    plt.plot(df_unstack2[continent], label=continent)

plt.legend()
plt.xlabel('Year')
plt.ylabel('Populations')
plt.grid()
plt.show()
```
